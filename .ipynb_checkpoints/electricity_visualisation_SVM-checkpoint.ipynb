#### calculate daily consumption in typical weeks

Depending on your location and energy usage, you should decide how to best divide your year in a useful fashion. I am picking some representative weeks from winter, summer and transitional period. 

```python
# Define the functions to select the typical weeks
def find_first_monday_in_month(year, month):
    """
    Find the date of the first Monday in a given year and month.

    Parameters:
    - year (int): The year.
    - month (int): The month (1-12).

    Returns:
    - pd.Timestamp: The timestamp of the first Monday in the specified month and year.
    """
    date = pd.Timestamp(year, month, 1)
    while date.dayofweek != 0:  # Monday is 0
        date += pd.Timedelta(days=1)
    return date

def create_typical_seasonal_dataframes(df, year):

    """
    Create seasonal dataframes based on the provided DataFrame and year.

    Parameters:
    - df (pd.DataFrame): The DataFrame containing the data.
    - year (int): The year for which seasonal dataframes are to be created.

    Returns:
    - tuple: A tuple containing three DataFrames:
        1. combined_winter_df (pd.DataFrame): DataFrame containing data for both winter periods.
        2. summer_df (pd.DataFrame): DataFrame containing data for the summer period.
        3. combined_transitional_df (pd.DataFrame): DataFrame containing data for both transitional periods.
    """
    
    # Define start and end dates for winter periods
    winter_start_date_1 = find_first_monday_in_month(year, 12)  # First Monday in December
    winter_end_date_1 = winter_start_date_1 + pd.DateOffset(days=20)  # Extend for the following 21 days

    # I am defining two distinct winter periods, omitting the Christmas Holidays
    # Additional winter period: First Monday after new year and the following 8 weeks
    winter_start_date_2 = find_first_monday_in_month(year + 1, 1)  # First Monday after new year (January)
    winter_end_date_2 = winter_start_date_2 + pd.DateOffset(weeks=8)  # 8 weeks after the start

    winter_df_1 = df[(df.index >= winter_start_date_1) & (df.index <= winter_end_date_1)]
    winter_df_2 = df[(df.index >= winter_start_date_2) & (df.index <= winter_end_date_2)]

    # Find the first Monday in June
    summer_start_date = find_first_monday_in_month(year, 6)

    # Calculate the end of summer, which is 16 weeks after the start
    summer_end_date = summer_start_date + pd.DateOffset(weeks=16)  # 16 weeks after the start

    # Find the first Monday in April
    transitional_start_date_1 = find_first_monday_in_month(year, 4)
    transitional_end_date_1 = transitional_start_date_1 + pd.DateOffset(weeks=8)  # 8 weeks after the start

    # Find the first Monday in October
    transitional_start_date_2 = find_first_monday_in_month(year, 10)
    transitional_end_date_2 = transitional_start_date_2 + pd.DateOffset(weeks=4)  # 4 weeks after the start

    transitional_df_1 = df[(df.index >= transitional_start_date_1) & (df.index <= transitional_end_date_1)]
    transitional_df_2 = df[(df.index >= transitional_start_date_2) & (df.index <= transitional_end_date_2)]
    
    combined_winter_df = pd.concat([winter_df_1, winter_df_2])
    combined_transitional_df = pd.concat([transitional_df_1, transitional_df_2])
    summer_df = df[(df.index >= summer_start_date) & (df.index <= summer_end_date)]

    return combined_winter_df, summer_df, combined_transitional_df
```

```python
# Initialize empty DataFrames for aggregation
winter_typical_df = pd.DataFrame()
summer_typical_df = pd.DataFrame()
transitional_typical_df = pd.DataFrame()

for year in range(2015, 2023):
        winter_df, summer_df, transitional_df = create_typical_seasonal_dataframes(raw_data, year)
        
        # Concatenate DataFrames for each season
        winter_typical_df = pd.concat([winter_typical_df, winter_df])
        summer_typical_df = pd.concat([summer_typical_df, summer_df])
        transitional_typical_df = pd.concat([transitional_typical_df, transitional_df])
```

```python
# Inspect the data
transitional_typical_df
```

Great! Now we have three different dataframes, each containing typical days of winter, summer and transitional weeks. In a next step, we want to average the weekdays from those representative weeks to see our typical usage patterns. 

```python
# Define a dictionary named seasonal_data where each key-value pair corresponds to a season (e.g., 'winter', 'summer', 'transitional') and its associated DataFrame
season_data_typical = {'winter': winter_typical_df, 'summer': summer_typical_df, 'transitional': transitional_typical_df}
```

```python
# Inspect the dictionary with the typical dataframes. 
season_data_typical
```

```python
# We will use the dataframes in season_data_typical to calculate the average weeks for summer, winter and transitional period
# Iterate over each season and its DataFrame and extract weekday, hour and minute information 
typical_weeks = {}  # Dictionary to store separate DataFrames for each season
for season, df in season_data_typical.items():
    df_copy = df.copy()  # Make a copy to avoid modifying the original DataFrame
    df_copy['weekday'] = df_copy.index.weekday  # Extract weekday from index
    df_copy['hour'] = df_copy.index.hour  # Extract hour from index
    df_copy['minute'] = df_copy.index.minute  # Extract minute from index
    typical_weeks[season] = df_copy  # Store separate DataFrame for the season
```

```python
typical_weeks
```

```python
# Group by weekday, hour, and minute, and calculate mean and confidence intervals
for season, df in typical_weeks.items():
    grouped = df.groupby(['weekday', 'hour', 'minute'])['value_measured']  # Group by weekday, hour, and minute
    # Calculate mean and confidence intervals for each group
    typical_weeks[season] = grouped.apply(lambda group: (
        group.mean(),
        group.mean() - 1.96 * group.std() / np.sqrt(len(group)),
        group.mean() + 1.96 * group.std() / np.sqrt(len(group)),
        group.sum()
    )).apply(pd.Series).rename(columns={0: 'mean', 1: 'ci_low', 2: 'ci_high', 3: 'sum'})

    # Reset index
    typical_weeks[season] = typical_weeks[season].reset_index()
```

```python
typical_weeks
```

As a last step, we want to map the numeric weekdays to the actual weekday names. 

```python
# Define a dictionary to map weekday numbers to weekday names
weekday_names = {
    0: 'Monday',
    1: 'Tuesday',
    2: 'Wednesday',
    3: 'Thursday',
    4: 'Friday',
    5: 'Saturday',
    6: 'Sunday'}

# Map weekdays in all result_dfs
for season, df in typical_weeks.items():
    df['weekday'] = df['weekday'].map(weekday_names)
```

```python
# Inspect the resulting dictionary with the average typical weeks. 
print(typical_weeks)
```

```python
# Plot the typical weeks along with the confidence intervals
import matplotlib.pyplot as plt

# Create a figure with three subplots
fig, axes = plt.subplots(nrows=1, ncols=3, figsize=(15, 5), sharey=True)

# Plot Row-wise Statistics for each season DataFrame with shaded area for standard deviation
for i, (season, df) in enumerate(typical_weeks.items()):
    axes[i].plot(df.index, df['mean'], label='Mean Usage')
    axes[i].fill_between(df.index,
                         df['mean'] - df['ci_low'],
                         df['mean'] + df['ci_high'],
                         alpha=0.2, label='Confidence Interval')
    axes[i].set_title(f"{season.capitalize()} DataFrame")
    axes[i].legend()
    # Set x-axis ticks to show only one label per 96 values
    axes[i].set_xticks(df.index[::96])
    axes[i].set_xticklabels(df['weekday'][::96], rotation=45, ha='right')

# Adjust layout to prevent overlapping
plt.tight_layout()

# Show the plot
plt.show()
```

Isnt this nice? You can already judge on which time and weekdays you are using the most energy.

#### define outlier days using support vector machines

After having seen some interesting features in the typical weeks, we want to go back to the raw data and find excepetional days in each season. For this, we are using one class support vector machines. First, we want to train the model on some typical days. We will use the dictionary we created before (it contains the typical weeks of the seasons, but the days were not averaged yet).

```python
from sklearn.svm import OneClassSVM
from numpy import where
```

```python
# We want to add some more features to the typical weeks
def iqr(series):
    return np.percentile(series, 75) - np.percentile(series, 25)

def process_dataframe_daily_for_SVMs(df):
    # Ensure index is a DateTimeIndex
    df.index = pd.to_datetime(df.index)
    
    # Grouping by date and calculating mean, sum, and standard deviation
    df_agg = df.groupby(df.index.date)['value_measured'].agg(['mean', 'sum', 'std', iqr])

    # Convert index to DateTimeIndex if not already
    df_agg.index = pd.to_datetime(df_agg.index)
    
    # Extract weekday information from the index
    df_agg['weekday'] = df_agg.index.dayofweek

    # Scale weekday between 0 and 50
    df_agg['scaled_weekday'] = 50 * df_agg['weekday'] / 6
    
    # Renaming columns for clarity
    df_agg.rename(columns={'mean': 'daily_mean', 
                          'sum': 'daily_sum', 
                          'std': 'daily_stdev',
                          'iqr': 'daily_iqr'}, inplace=True)
    
    # Now we can drop the weekday column
    df_agg.drop(columns=['weekday'], inplace=True)  # Drop 'weekday' and 'hour' columns

    # Last, we need to delete all rows containing nans
    df_agg.dropna(inplace=True)
    
    return df_agg
```

```python
# The dictionary "season_data_typical" contains the representative weeks (not yet averaged)
season_data_typical
```

```python
# Initialize dictionary to store the processed dataframes
season_data_typical_processed = {}

for season, df in season_data_typical.items():
    # Process each dataframe
    df_typical_processed = process_dataframe_daily_for_SVMs(df)
    # Store the processed dataframe in the dictionary
    season_data_typical_processed[season] = df_typical_processed
```

```python
season_data_typical_processed
```

Now we can train the models on the samples we just generated. 

```python
# Initialize a dictionary to store the models
season_models = {}

# Train a OneClassSVM model for each season's dataframe
for season, df in season_data_typical_processed.items():

    # The gamma parameter specifies the kernel coefficient for the Radial Basis Function (RBF) kernel
    # Smaller value of gamma makes the decision boundary smoother, 
    # Larger value makes it more wiggly, potentially leading to overfitting
    model = OneClassSVM(kernel='rbf', gamma=0.001, nu=0.003).fit(df)
    season_models[season] = model
```

```python
season_models['winter']
```

After fitting the models to the typical weeks, we will use all the data in the dataframe we have and identify those, that differ markedly from the models we defined. 

```python
# Function to divide the raw data according to the seasons based on months
def divide_seasons(raw_data):

    """
    Divide the raw data into seasons based on months.

    Parameters:
    raw_data (pd.DataFrame): DataFrame containing the raw data with a datetime index.

    Returns:
    dict: A dictionary containing DataFrames divided into different seasons.
          Keys represent the seasons ('summer', 'winter', 'transitional').
    """ 
    
    
    # Ensure the index is a datetime index
    raw_data.index = pd.to_datetime(raw_data.index)
    
    # Filter data for each season based on months
    summer = raw_data[raw_data.index.month.isin([6, 7, 8])]
    winter = raw_data[raw_data.index.month.isin([12, 1, 2])]
    
    # Filter data for the transitional period (spring and autumn)
    transitional = raw_data[~raw_data.index.month.isin([6, 7, 8, 12, 1, 2])]
    
    # Create a dictionary to store the resulting dataframes
    seasonal_data = {
        'summer': summer,
        'winter': winter,
        'transitional': transitional
    }
    
    return seasonal_data
```

```python
# Assuming raw_data is your original DataFrame with a datetime index
seasonal_data_all = divide_seasons(raw_data)
```

```python
# Inspect the result
print(seasonal_data_all['winter'])
```

Now, we can use the raw data - divided into the seasons - and use the models created to predict outlier days. First, we need to also process the seasonal_data_all.

```python
# Initialize dictionary to store the processed raw data, that were previously divided into the seasons
season_data_all_processed = {}

for season, df in seasonal_data_all.items():
    # Process each dataframe
    df_typical_processed = process_dataframe_daily_for_SVMs(df)
    # Store the processed dataframe in the dictionary
    season_data_all_processed[season] = df_typical_processed
```

```python
# Initialize a dictionary to store the predictions. 
predictions = {}

# We are using the models we trained on the typical weeks and the method "predict". 
for season in season_data_all_processed.keys():
    model = season_models[season]  # Get the model for the current season
    data = season_data_all_processed[season]  # Get the data for the current season
    predictions[season] = model.predict(data)  # Make predictions for the current season
```

```python
predictions['summer']
```

The dictionary "predictions" now contains the results. "-1" means, that an outlier was detected. 

```python
# We want to filter the rows that contain outliers ("-1").
def filter_outliers_indices(y_pred ):
    outlier_indices = np.where(y_pred == -1)[0]  # filter outlier indices
    return outlier_indices
```

```python
# Create lists to store the dates of outliers for each season
all_winter_outlier_dates = []
all_summer_outlier_dates = []
all_transitional_outlier_dates = []

# Iterate over seasons and filter outlier indices
for season, y_pred_season in predictions.items():
    outlier_indices = filter_outliers_indices(y_pred_season)
    season_data = season_data_all_processed[season]
    outlier_dates = [season_data.index[index] for index in outlier_indices]

    # Append outlier dates to the corresponding list
    if season == 'winter':
        all_winter_outlier_dates.extend(outlier_dates)
    elif season == 'summer':
        all_summer_outlier_dates.extend(outlier_dates)
    elif season == 'transitional':
        all_transitional_outlier_dates.extend(outlier_dates)
```

```python
all_transitional_outlier_dates
```

```python
import matplotlib.pyplot as plt

# We might not want to plot all the data, so here we can specify the start and end dates.
start_date = '2019-11-01'
end_date = '2019-12-31'

# Assuming the index of the DataFrames is of datetime type, if not, convert it
#seasonal_data_all['winter'].index = pd.to_datetime(seasonal_data_all['winter'].index)
#seasonal_data_all['summer'].index = pd.to_datetime(seasonal_data_all['summer'].index)
#seasonal_data_all['transitional'].index = pd.to_datetime(seasonal_data_all['transitional'].index)

# Filter data within the date range
winter_filtered = seasonal_data_all['winter'][start_date:end_date]
summer_filtered = seasonal_data_all['summer'][start_date:end_date]
transitional_filtered = seasonal_data_all['transitional'][start_date:end_date]

# Filter outlier dates within the date range
winter_outliers_within_range = [date_range for date_range in all_winter_outlier_dates if date_range <= pd.Timestamp(end_date) and date_range >= pd.Timestamp(start_date)]
summer_outliers_within_range = [date_range for date_range in all_summer_outlier_dates if date_range <= pd.Timestamp(end_date) and date_range >= pd.Timestamp(start_date)]
transitional_outliers_within_range = [date_range for date_range in all_transitional_outlier_dates if date_range <= pd.Timestamp(end_date) and date_range >= pd.Timestamp(start_date)]

fig, ax = plt.subplots(figsize=(12, 4))

# Plot the filtered data with dots
ax.plot(winter_filtered.index, winter_filtered, 'o', markersize=5, color='blue', label='Winter')
ax.plot(summer_filtered.index, summer_filtered, 'o', markersize=5, color='red', label='Summer')
ax.plot(transitional_filtered.index, transitional_filtered, 'o', markersize=5, color='green', label='Transitional')

# Adding shaded areas for outlier dates within the range
for outlier_dates, color in zip([winter_outliers_within_range, summer_outliers_within_range, transitional_outliers_within_range],
                                ['blue', 'red', 'green']):
    for date_range in outlier_dates:
        start, end = date_range, date_range + pd.Timedelta(days=1)
        ax.axvspan(max(start, pd.Timestamp(start_date)), min(end, pd.Timestamp(end_date)), color=color, alpha=0.3)

# Adding labels and title
ax.set_xlabel('Time')
ax.set_ylabel('Value')
ax.set_title('Seasonal Data within Date Range')
ax.legend()

# Display the plot
plt.show()
```
